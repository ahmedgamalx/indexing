```python
import requests
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

url = "https://en.wikipedia.org/wiki/Data_science"
response = requests.get(url)

if response.status_code == 200:
    content = response.text

    with open("data_science.txt", "w", encoding="utf-8") as file:
        file.write(content)
        print("Data saved successfully.")
else:
    print("Failed to retrieve data from the URL.")
```

```python
def remove_head_tags(text):
    pattern = r">(?:[a-zA-Z0-9 -\"?\s]+)<\/"
    return re.findall(pattern, text)
```

```python
data=remove_head_tags(content)
```

```python
def remove_prefix(lst):
    return [element[1:] for element in lst]
```

```python
data=remove_prefix(data)
```

```python
def remove_prefixx(lst):
    return [element[:-2] for element in lst]
```

```python
data=remove_prefixx(data)
```

```python
def split_list_elements_if_multi_word(list):

  new_list = []

  for word in list:
    if len(word.split()) > 1:
      new_list.extend(word.split())
    else:
      new_list.append(word)

  return new_list
```

```python
def count_repetition(list1):
  counts = {}
  for item in list1:
    if item in counts:
      counts[item] += 1
    else:
      counts[item] = 1
  return counts



```
